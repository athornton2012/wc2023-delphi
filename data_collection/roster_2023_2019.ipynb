```python
def get_roster(country):
    url = 'https://www.foxsports.com/soccer/'+str(country)+'-women-team-roster'
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    tables = soup.find_all('div', class_ = 'secondary-nav-content-container')
    html_content = str(tables)
    df = pd.read_html(html_content)
    return df
```

```python
def clean_df(df):
    roster_dict = {}
    positions = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
    for i in range(4):
        position = positions[i]
        position_df = df[i]
        position_df.rename(columns = {position.upper():'Player'}, inplace = True)
        roster_dict[position] = position_df
    df_new = pd.concat(roster_dict.values(), keys=roster_dict.keys()).reset_index()
    df_new['Jersey'] = df_new['Player'].apply(lambda x: x.split('#')[-1])
    df_new['Player'] = df_new['Player'].apply(lambda x : x.split('#')[0])
    df_new['Player'] = df_new['Player'].apply(lambda x: x.strip() if isinstance(x, str) else x)
    df_new.drop(['level_1','POS'], axis = 1, inplace = True)
    df_new.rename(columns = {'level_0':'Position'}, inplace = True)
    return df_new
```

```python
def get_diff(df_2019, df_2023):
    df_diff = df_2023.copy()
    df_diff['2019 Roster'] = df_2023['Player'].apply(lambda x: 'Yes' if x in df_2019['Player'].values else 'No')
    df_diff.rename(columns = {'AGE':'Age', 'HT': 'Height', 'WT':'Weight'}, inplace = True)
    df_diff= df_diff[['Player','Position',  'Age', 'Height', 'Weight', 'Jersey', '2019 Roster']]
    return df_diff
```

```python
def roster_color(value):
    if value == 'Yes':
        return 'background-color: lavender; color: black'
    elif value == 'No':
        return 'background-color: ghostwhite; color: black'
    else:
        return ''
```

```python
def position_color(value):
    if value == 'Goalkeeper':
        return 'background-color: azure; color: black'
    elif value == 'Defender':
        return 'background-color: honeydew; color: black'
    elif value == 'Midfielder':
        return 'background-color: ivory; color: black'
    elif value == 'Forward':
        return 'background-color: oldlace; color: black'    
    else:
        return ''
```

```python
def get_style(df):
    df_styler = df.style. \
    set_properties(**{'background-color': 'white'}). \
    applymap(roster_color, subset=['2019 Roster']). \
    applymap(position_color, subset=['Position']).hide(axis='index'). \
    set_properties(**{'text-align': 'center'}, subset=['2019 Roster']). \
    set_table_styles([{'selector': 'table', 'props': [('border-collapse', 'separate'),('border-spacing', '10px')]}])
    display(df_styler)
```
