```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
sns.set()
```

```python
rating23_2 = pd.read_csv('../Datasets/FIFA/FIFA23_rating/rating_23.csv')
rating23 = pd.read_csv('../Datasets/FIFA/FIFA23_rating/rating_23_Feb.csv')
rating22 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_22.csv')
rating21 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_21.csv')
rating20 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_20.csv')
rating19 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_19.csv')
rating18 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_18.csv')
rating17 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_17.csv')
rating16 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_16.csv')
```

```python
league_1 = rating23.leaguename.value_counts()
```

```python
rating23_2.head()
```

```python
fig, ax = plt.subplots()
ax.pie(league_1.values, labels=league_1.index, autopct='%1.1f%%')
ax.set_title('Number of rated player per league in FIFA Feb release',
             loc ='left')
plt.savefig('../imgs/FIFA_23/leagues_Feb.png',bbox_inches='tight')
```

```python
league_2 = rating23_2.leaguename.value_counts()
```

```python
fig, ax = plt.subplots()
ax.pie(league_2.values, labels=league_2.index, autopct='%1.1f%%')
ax.set_title('Number of rated player per league in FIFA March release',
             loc ='left')
plt.savefig('../imgs/FIFA_23/leagues_Mar.png',bbox_inches='tight')
```

```python
team_1 = rating23.teamname.value_counts()
```

```python
team_1
```

```python
fig, ax = plt.subplots(figsize =(16, 12))
ax.barh(team_1.index, team_1.values, color = 'maroon', height = 0.4)
ax.invert_yaxis()
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
ax.set_title('Number of rated player per club/national team in FIFA Feb release',
             loc ='left')
plt.savefig('../imgs/FIFA_23/team_Feb.png',bbox_inches='tight')
```
