```python
Group_1 = ['FIFA World Cup', 'Euro']
Group_2 = ['OFC Championship','African Championship', 'CONCACAF Championship', 
          'Olympic Games', 'AFC Asian Cup']
Group_3 = ['Friendly']
```

```python
def k_value(tournament):

    if tournament in Group_1:
        k=60
    elif tournament in Group_2:
        k=40
    elif tournament in Group_3:
        k=20
    else:
        k=30
    return k
```

```python
def expected_result(elo_h, elo_a):
    dr=elo_h-elo_a
    we=(1/(10**(-dr/400)+1))
    return [np.round(we,3),1-np.round(we,3)]
```

```python
def actual_result(home_score,away_score):
    if home_score>away_score:
        w_h=1
        w_a=0
    elif home_score==away_score:
        w_h=0.5
        w_a=0.5
    elif home_score<away_score:
        w_h=0
        w_a=1
    return [w_h,w_a]
```

```python
def calculate_elo(elo_h,elo_a,home_score,away_score,tournament):
    
    k=k_value(tournament)
    w_h,w_a=actual_result(home_score,away_score)
    we_h,we_a=expected_result(elo_h,elo_a)

    elo_h_n=elo_h+k*(w_h-we_h)
    elo_a_n=elo_a+k*(w_a-we_a)

    return elo_h_n,elo_a_n
```

```python
results_elo = results_new.copy()
```

```python
current_elo = {}
```

```python
for idx, row in results_new.iterrows():
    
    home = row['home_team']
    away = row['away_team']
    home_score = row['home_score']
    away_score = row['away_score']
    tournament = row['tournament']
    neutral = row['neutral']
    
    if home in current_elo.keys() and away in current_elo.keys():
        if home_score>away_score:
            results_elo.loc[idx, 'home_win'] = 2
        elif home_score==away_score:
            results_elo.loc[idx, 'home_win'] = 1
        elif home_score<away_score:
            results_elo.loc[idx, 'home_win'] = 0            
    
    if home not in current_elo.keys():
        current_elo[home] = 1300
    if away not in current_elo.keys():
        current_elo[away] = 1300
    
    if neutral == False:
        elo_h = current_elo[home] +100
    else:
        elo_h = current_elo[home]
    
    elo_a = current_elo[away]
    elo_h_n, elo_a_n = calculate_elo(elo_h, elo_a, home_score, away_score, tournament)
    
    current_elo[home] = elo_h_n
    current_elo[away] = elo_a_n
    
    results_elo.loc[idx, 'elo_h'] = elo_h
    results_elo.loc[idx, 'elo_a'] = elo_a
    
```

```python
current_elo_df = pd.DataFrame.from_dict(current_elo,orient='index', columns=['Elo'])
```

```python
def count_frequency(x, col1, col2, idx):
    frequency1 = col1[:idx].eq(x).sum()
    frequency2 = col2[:idx].eq(x).sum()
    return frequency1 + frequency2
```

```python
results_elo['home_experience'] = results_elo.apply(lambda row: count_frequency(
    row['home_team'], results_elo['home_team'], results_elo['away_team'], row.name), axis=1)
```

```python
results_elo['away_experience'] = results_elo.apply(lambda row: count_frequency(
    row['away_team'], results_elo['home_team'], results_elo['away_team'], row.name), axis=1)
```

```python
results_elo['elo_diff'] = results_elo['elo_h'] - results_elo['elo_a']
```

```python
subset_below = df_1999[df_1999['elo_h'] < 2315]
subset_above = df_1999[df_1999['elo_h'] > 2315]

# Create two separate plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

# Define the logistic function
def logistic_function(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Plot for elo_h < 2315
scatter1 = ax1.scatter(subset_below['elo_diff'], subset_below['home_win'], s=4)
ax1.set_xlabel('Elo Difference (elo_h - elo_a)')
ax1.set_ylabel('home_win')
ax1.set_title('Elo_h < 2315')

# Fit logistic regression curve for elo_h < 2315
x1 = subset_below['elo_diff']
y1 = subset_below['home_win']
popt1, pcov1 = curve_fit(logistic_function, x1, y1)
x_range1 = np.linspace(min(x1), max(x1), 100)
ax1.plot(x_range1, logistic_function(x_range1, *popt1), color='red')

# Plot for elo_h > 2315
scatter2 = ax2.scatter(subset_above['elo_diff'], subset_above['home_win'], s=4)
ax2.set_xlabel('Elo Difference (elo_h - elo_a)')
ax2.set_ylabel('home_win')
ax2.set_title('Elo_h > 2315')

# Fit logistic regression curve for elo_h > 2315
x2 = subset_above['elo_diff']
y2 = subset_above['home_win']
popt2, pcov2 = curve_fit(logistic_function, x2, y2)
x_range2 = np.linspace(min(x2), max(x2), 100)
ax2.plot(x_range2, logistic_function(x_range2, *popt2), color='red')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
