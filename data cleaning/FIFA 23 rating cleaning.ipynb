```python
import numpy as np
import pandas as pd
```

```python
rating = pd.read_csv('../Datasets/FIFA/FIFA23_rating/player_rating.csv')
```

```python
rating.head()
```

```python
na_s = rating.isnull().sum()
```

```python
col_0 = (rating == 0).sum(axis = 0)
```

```python
drop_list = [i for i in na_s.index if na_s[i] > 400]
```

```python
drop_list
```

```python
for i in col_0.index:
    if col_0[i]>400:
        drop_list.append(i)
```

```python
drop_list
```

```python
rating.drop(columns = drop_list, inplace = True)
```
