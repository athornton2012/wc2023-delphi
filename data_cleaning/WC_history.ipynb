```python
import numpy as np
import pandas as pd
```

```python
WC_1991 = pd.read_csv('../Datasets/FBRef/History/1991.csv', index_col = 0)
WC_1995 = pd.read_csv('../Datasets/FBRef/History/1995.csv', index_col = 0)
WC_1999 = pd.read_csv('../Datasets/FBRef/History/1999.csv', index_col = 0)
WC_2003 = pd.read_csv('../Datasets/FBRef/History/2003.csv', index_col = 0)
WC_2007 = pd.read_csv('../Datasets/FBRef/History/2007.csv', index_col = 0)
WC_2011 = pd.read_csv('../Datasets/FBRef/History/2011.csv', index_col = 0)
WC_2015 = pd.read_csv('../Datasets/FBRef/History/2015.csv', index_col = 0)
WC_2019 = pd.read_csv('../Datasets/FBRef/History/2019.csv', index_col = 0)
```

```python
def cleanup(df):
    df = df.dropna(subset = ['Rk'])
    df['Squad'] = df['Squad'].apply(lambda x: ''.join(x.split()[1:]))
    df = df.drop(columns = 'Notes')
    return df
```

```python
WC_1991 = cleanup(WC_1991)
WC_1995 = cleanup(WC_1995)
WC_1999 = cleanup(WC_1999)
WC_2003 = cleanup(WC_2003)
WC_2007 = cleanup(WC_2007)
WC_2011 = cleanup(WC_2011)
WC_2015 = cleanup(WC_2015)
WC_2019 = cleanup(WC_2019)
```

```python
WC_history = pd.concat([WC_1991.assign(source='1991'), WC_1995.assign(source = '1995'),
                      WC_1999.assign(source='1999'), WC_2003.assign(source = '2003'),
                      WC_2007.assign(source = '2007'), WC_2011.assign(source = '2011'),
                      WC_2015.assign(source = '2015'), WC_2019.assign(source = '2019')])
```

```python
WC_history = WC_history.rename(columns = {'source': 'Year'})
```
