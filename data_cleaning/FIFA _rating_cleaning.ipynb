```python
rating23_2 = pd.read_csv('../Datasets/FIFA/FIFA23_rating/rating_23.csv')
rating23 = pd.read_csv('../Datasets/FIFA/FIFA23_rating/rating_23_Feb.csv')
rating22 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_22.csv')
rating21 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_21.csv')
rating20 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_20.csv')
rating19 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_19.csv')
rating18 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_18.csv')
rating17 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_17.csv')
rating16 = pd.read_csv('../Datasets/FIFA/FIFA22_rating/female_players_16.csv')
```

```python
def drop_1(dataset):
    '''drop columns with more than 200 missing data or more than 200 values = 0'''
    na_s = rating23.isnull().sum()
    col_0 = (rating23 == 0).sum(axis = 0)
    drop_list = [i for i in na_s.index if na_s[i] > 200]
    for i in col_0.index:
        if col_0[i]>200:
            drop_list.append(i)
    dataset.drop(columns = drop_list, inplace = True)
```

```python
drop_1(rating23_2)
drop_1(rating23)
drop_1(rating22)
drop_1(rating21)
drop_1(rating20)
drop_1(rating19)
drop_1(rating18)
drop_1(rating17)
drop_1(rating16)
```

```python
rating23_2_new = rating23_2[new_col_1]
rating23_2_new['birthdate']=rating23_2_new.apply(lambda x:x['birthdate'].split()[0], axis=1)

rating23_2_new['birthdate'] = rating23_2_new.apply(lambda x:datetime.strptime(x['birthdate'], '%m/%d/%Y').year, axis =1)

rating23_2_new = rating23_2_new.rename({'fullNameForSearch':'full_name', 'birthdate':'birthyear', 'preferredposition1label': 'position'}, axis=1)
rating23_2_new.rename(columns = {'overall_rating':'overall','pac':'pace', 'sho':'shooting', 'pas': 'passing', 
                               'dri':'dribbling','def':'defending'}, inplace = True)
```

```python
def get_nation(df):
    nation_list = df.nationality.unique()
    nation_dict = {}
    for num in nation_list:
        r = requests.get('https://www.ea.com/games/fifa/fifa-23/womens-ratings/player-filter/player-nationality/'+str(num))
        soup = BeautifulSoup(r.content, 'html.parser')
        title = soup.find('title')
        nation_dict[num] = title.text.split()[0]
    df['nationality'] = df['nationality'].replace(nation_dict)
    return df
```

```python
rating23_new = get_nation(rating23_new)
```

```python
rating23_2_new = get_nation(rating23_2_new)
```
